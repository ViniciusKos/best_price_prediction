# Data Cleaning

```python
import pandas as pd
import numpy as np
from datetime import datetime
import re
pd.set_option('display.max_columns', 100)
```

```python
data = pd.read_csv("df_products.csv")
data.head(2)
```

```python
data2 = data.copy()


for i in ['product_name', 'fit', 'description', 'material']:
    data2[i] = data2[i].apply( lambda x: x.replace(" ", "_").lower())
    data2[i] = data2[i].apply( lambda x: x.replace("/", "_").lower())

data2['composition'] = data2['composition'].str.lower()

data2['price'] = data2['price'].apply( lambda x: x.replace("$", "")).astype(float)

data2['scrapy_datetime'] = pd.to_datetime(data2['scrapy_datetime'], format = "%Y-%m-%d %H:%M:%S")

# data2['size_number'] = data2['size'].apply( lambda x: re.search("\d+[^cm]*", x).group(0) if pd.notnull( x ) else x ).astype('float').astype('float64')

# data2['size_model'] = data2['size'].str.extract( '(\d+\/\d+)' )

data2 = data2.drop(['size'], axis=1)

for i in ['style_id', 'color_id', 'product_id']:
    data2[i] = data2[i].astype(int)


print(data2.shape)

#for i in ['lining', 'shell', 'pocket']:
#    data2 = data2[~data2['composition'].str.contains( i, na=False) ]

print(data2.shape)
data2 = data2.drop_duplicates()
print(data2.shape)


data2 = data2.reset_index(drop=True)

print(data2.shape)

data2 = data2[['']]

data2.head(2)
```

```python
data2 = data2[[ 'product_id','style_id', 'color_id', 'product_name', 'composition', 'color', 'fit','price' ]]
data2.head(3)
```

```python
data2.to_csv( "df_products_cleaned.csv", index=False )
```
